```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import csv
```

```python
from datetime import date
```

```python
def cap_Embalses (comunidad, lista_Embalses):
    page2 = requests.get(comunidad)         
    soup2 = BeautifulSoup(page2.content)
    tabla = soup2.find_all('table')[1]  #en la web por comunidad hay varias tablas; nos interesa la segunda
    #Definimos las variabbles
    Embalse=""
    Capacidad=""
    Embalsada=""
    Variacion=""
    fecha=date.today().isoformat() # guardamos los datos en  forma AAAA-MM-DD
    nroFila=1
    for fila in tabla.find_all("tr", class_="ResultadoCampo"): #buscamos los tr que tengan clase "ResultadoCampo"
            if nroFila==1:
                nroCelda=0
                for celda in fila.find_all('td'):
                    if nroCelda==0:
                        Embalse=celda.text
                        Embalse=Embalse.strip('\n') # eliminamos todos los saltos de línea
                    if nroCelda==1:
                        Capacidad=celda.text
                    if nroCelda==2:
                        Embalsada=celda.text
                    if nroCelda==3:
                        Variacion=celda.text
                    nroCelda=nroCelda+1
                DatosEmbalse = [Embalse,Capacidad,Embalsada, Variacion, fecha] 
                lista_Embalses.append (DatosEmbalse) #añadimos los datos a la lista 
    nroFila=nroFila+1
    return
```

```python
lista_Embalses =[] #definimos la lista donde almacenaros los valores extraídos
```

```python
lista_Embalses_cabecera=["Embalse","Capacidad", "Emblasada","Variación","Fecha datos"] 
```

```python
lista_Embalses.append(lista_Embalses_cabecera)  #añadimos la cabecera a la lista
```

```python
page = requests.get("https://www.embalses.net") 
```

```python
soup = BeautifulSoup(page.content)
```

```python
#aquí vamos link a link incluído en todos los links, pero sólo incluímos los que en su cabecera http incluyan
#el texto "comunidad-"
for link in soup.find_all('a'):
    #print (link.get('href'))
    web_Com = link.get('href')
    if (type(web_Com)is str):
        if 'comunidad-' in web_Com:
            print ("Extrayendo datos de", web_Com)
            cap_Embalses (web_Com, lista_Embalses)  #pasamos como parámetros la web y la lista de valores
    
```

```python
#el archivo agua_embalsada.csv estará en el mismo directorio que el jupyter notebook; si quisieramos cambiarlo,
#lo indicaríamos aquí. Al abrir el fichero con 'a' lo que hacemos es añadir al fichero las nuevas líneas
with open('agua_embalsada.csv', 'a', newline="") as csv_file: # con newline="" evitamos los saltos de línea
    writer = csv.writer(csv_file)
    for embalse in lista_Embalses :
        writer.writerow(embalse)
```
